```python
import sys
import time
import os
import xarm

sys.path.append("../")
from Drivers.pipette_driver import Pipette

```

```python
time.sleep(10)
ip = '192.168.1.240'

arm = xarm.XArmAPI(ip, do_not_open=True)

arm.connect()

arm.motion_enable(enable=True)
arm.set_mode(0)
arm.set_state(state=0)
arm.set_gripper_enable(True)
arm.set_gripper_mode(0)

```

```python
pipette = Pipette("COM6")
pipette.initiate_rline()
```

```python
def PickupPipette():
    arm.set_gripper_position(800, wait=True)
    arm.set_position(x=200, y=700, z=190, roll=0, pitch=90, yaw=90, speed=40, wait=True)
    arm.set_position(x=55, y=700, z=190, roll=0, pitch=90, yaw=90, speed=40, wait=True)
    arm.set_position(x=60, y=750, z=183, roll=0, pitch=90, yaw=90, speed=40, wait=True)
    arm.set_gripper_position(480, wait=True)
    arm.set_position(x=60, y=600, z=183, roll=0, pitch=90, yaw=90, speed=40, wait=True)
```

```python
def GoToVialOne():
    arm.set_position(x=300, y=400, z=400, roll=0, pitch=90, yaw=0, speed=20, wait=True)
    time.sleep(5)
    arm.set_position(x=370, y=470, z=400, roll=0, pitch=90, yaw=0, speed=20, wait=True)
    arm.set_position(x=373, y=470, z=310, roll=0, pitch=90, yaw=0, speed=20, wait=True)
    
```

```python
def VialOneToTwo():
    arm.set_position(x=373, y=470, z=380, roll=0, pitch=90, yaw=0, speed=20, wait=True)
    arm.set_position(x=373, y=535, z=380, roll=0, pitch=90, yaw=0, speed=20, wait=True)
    arm.set_position(x=373, y=535, z=310, roll=0, pitch=90, yaw=0, speed=20, wait=True)
```

```python
def VialTwoToThree():
    arm.set_position(x=373, y=535, z=380, roll=0, pitch=90, yaw=0, speed=20, wait=True)
    arm.set_position(x=377, y=600, z=380, roll=0, pitch=90, yaw=0, speed=20, wait=True)
    arm.set_position(x=377, y=600, z=330, roll=0, pitch=90, yaw=0, speed=20, wait=True)
```

```python
def GoToInitialPosition(s):
    arm.set_position(x=400, y=0, z=133, roll=0, pitch=90, yaw=0, speed=s, wait=True)
```

```python
#DO NOT RUN WHILE HOLDING PIPETTE
#GoToInitialPosition(100)
```

```python
#above vial one
arm.set_position(x=373, y=470, z=400, roll=0, pitch=90, yaw=0, speed=20, wait=True)
```

```python
#From above vial one moving liquid to vial two then ejecting tip in vial three (good for calibration testing)
arm.set_position(x=373, y=470, z=400, roll=0, pitch=90, yaw=0, speed=20, wait=True)
arm.set_position(x=373, y=470, z=310, roll=0, pitch=90, yaw=0, speed=20, wait=True)
pipette.aspirate(2000) #aspirating 2000ul
time.sleep(5)
VialOneToTwo()
pipette.dispense(2000)
time.sleep(5)
pipette.blowout()
VialTwoToThree()
pipette.eject_and_home()
```

```python
#The full sequence from picking up the pipette to dispensing water and ejecting the tip in vial three
PickupPipette()
GoToVialOne()
pipette.aspirate(2000)
time.sleep(5)
VialOneToTwo()
pipette.dispense(2000)
time.sleep(5)
pipette.blowout()
VialTwoToThree()
pipette.eject()
```

```python
#returning the pipette then Inital position
arm.set_position(x=60, y=600, z=183, roll=0, pitch=90, yaw=90, speed=40, wait=True)
arm.set_position(x=60, y=750, z=183, roll=0, pitch=90, yaw=90, speed=40, wait=True)
arm.set_gripper_position(800, wait=True)
arm.set_position(x=60, y=600, z=183, roll=0, pitch=90, yaw=90, speed=40, wait=True)
arm.set_position(x=60, y=600, z=183, roll=0, pitch=90, yaw=90, speed=40, wait=True)
arm.set_position(x=200, y=700, z=190, roll=0, pitch=90, yaw=90, speed=40, wait=True)
GoToInitialPosition(100)

```
