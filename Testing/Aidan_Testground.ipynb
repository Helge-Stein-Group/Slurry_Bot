```python
import time
import random
#place_funnel() 
scale.tare()
previous_weight = 0
last_scoop = None


## Initial spatula selection based on target weight
def select_spatula(weight):
    if weight < 1.0:
        return pickup_medium(), "Medium Spatula"
    else:
        return pickup_large(), "Large Spatula"

target_weight = float(input("Enter total target weight (g): "))

pickup_function, tool_name = select_spatula(target_weight)
print(f"Selected tool: {tool_name}")
## Initial defining of terms and weight calculated
weight_dispensed = scale.measure_stable().value
print(f"Initial weight dispensed: {weight_dispensed}")  
def calc_weight_needed(weight_dispensed, target_weight):
    weight_needed = target_weight - weight_dispensed
    return weight_needed
weight_needed = calc_weight_needed(weight_dispensed, target_weight)
print(f"Needed: {weight_needed}")  

## LOGIC FOR MORE THAN 1G HERE
## First scoop selected purely on weight needed
if target_weight >= 1.0:
    #### BIG SPATULA NEEDED ###
    if 1 <= target_weight < 1.5:
        scoop_big_right()
        last_scoop = scoop_big_right
    elif 1.5 <= target_weight < 2:
        scoop_big_left()
        last_scoop = scoop_big_left
    elif target_weight >= 2:
        scoop_big_middle()
        last_scoop = scoop_big_middle

    for _ in range(30):
        raw_measurement = scale.measure()
        if raw_measurement.stable and abs(raw_measurement.value - previous_weight) > 0.01:
            break
        time.sleep(0.1)
## Measure scoop 1, calculate percentage of mass required that was dispensed in scoop 1
    weight_dispensed = raw_measurement.value
    print(f"Weight after first scoop: {weight_dispensed} grams")
    weight_needed = calc_weight_needed(weight_dispensed, target_weight)
    print(f"Current weight needed: {weight_needed} grams")
    previous_weight = weight_dispensed
    fraction_left = weight_needed / target_weight
    print(f"Fraction remaining: {fraction_left:.4f}")
## Loop until 37% or less of mass required still left
## Helps stop scooping in same place over and over again getting 0, also helps negate effects of a differently fulled sample box (ie a super full one and a super empty will need different combinations)
    # LOOP until fraction_left < 0.37
    while fraction_left > 0.37:
        if fraction_left > 0.62:
            print('Rescooping big middle')
            scoop_big_middle()
            last_scoop = scoop_big_middle
            ## Last scoop logic because for two scoops in the same place in a row, the second will be half of the first, allows for more tailored selection
        elif 0.5 <= fraction_left <= 0.62:
            print('Rescooping big left')
            scoop_big_left()
            last_scoop = scoop_big_left
        elif 0.4 <= fraction_left < 0.5:
            print('Rescooping last scoop')
            if last_scoop:
                last_scoop()
            else:
                print("No last scoop remembered")
        elif 0.37 <= fraction_left < 0.4:
            print('Rescooping big right')
            scoop_big_right()
            last_scoop = scoop_big_right
        else:
            print('Unexpected state in loop')
            break

        for _ in range(30):
            raw_measurement = scale.measure()
            if raw_measurement.stable and abs(raw_measurement.value - previous_weight) > 0.01:
                break
            time.sleep(0.1)

        weight_dispensed = raw_measurement.value
        weight_needed = calc_weight_needed(weight_dispensed, target_weight)
        previous_weight = weight_dispensed
        fraction_left = weight_needed / target_weight

        print(f"Weight now: {weight_dispensed:.4f}g")
        print(f"Still needed: {weight_needed:.4f}g")
        print(f"Fraction remaining: {fraction_left:.4f}")

    ## Only enter medium scooping phase if within correct fraction range
    if 0.03 <= fraction_left < 0.37:
        print('Switching to small spatula')
        replace_big_spatula()
        pickup_medium()

        # Medium spatula logic based on weight needed
        while weight_needed > 0.05:
            if weight_needed >= 0.6:
                print("Using medium middle scoop")
                scoop_medium_middle()
            elif 0.4 <= weight_needed < 0.6:
                print("Using medium left scoop")
                scoop_medium_left()
            elif 0.3 <= weight_needed < 0.4:
                print("Using medium right scoop")
                scoop_medium_right()
            elif 0.2 <= weight_needed < 0.3:
                print("Using medium far left scoop")
                scoop_medium_farleft()
            elif 0.08 <= weight_needed < 0.2:
                print("Using micro scoop")
                scoop_tiny_right()
            elif 0.04 <= weight_needed < 0.08:
                print("Using smallest micro scoop")
                scoop_tiny_left()
            else:
                print("Remaining weight too small — consider nudging or final adjustment")

            for _ in range(30):
                raw_measurement = scale.measure()
                if raw_measurement.stable and abs(raw_measurement.value - previous_weight) > 0.01:
                    break
                time.sleep(0.1)

            weight_dispensed = raw_measurement.value
            print(f"Weight after scoop: {weight_dispensed} grams")
            weight_needed = calc_weight_needed(weight_dispensed, target_weight)
            print(f"Current weight needed: {weight_needed} grams")
            previous_weight = weight_dispensed

        # Final result check
# Final result check
# Final result check
        difference = target_weight - weight_dispensed  # signed difference

        if difference < -0.02:
            print("⚠️ Overshot the target.")
            print("Final weight: {:.4f}g (Overshoot by {:.4f}g)".format(weight_dispensed, abs(difference)))
            replace_medium_spatula()
        elif -0.02 <= difference <= 0.02:
            print("✅ Target weight reached.")
            print("Final weight: {:.4f}g (Off by {:.4f}g)".format(weight_dispensed, abs(difference)))
            replace_medium_spatula()
        else:
            print("❓ Loop ended unexpectedly — undershot.")
            print("Final weight: {:.4f}g (Still need {:.4f}g to reach target)".format(weight_dispensed, difference))



    else:
        print('Error — either less than 3% remaining or overshot too early')
## SMALL CASES LESS THA A GRAM
else:
    print("Using medium spatula")
    ## Slightly different logic, no percentages here yet as too much variance based on fullness of sample container
    ## This is much more severe here as compared to larger amounts, perhaps using scoops with more of the 'emptying' steps above the sample could help?
    while weight_needed > 0.02:
            if weight_needed >= 0.6:
                print("Using medium middle scoop")
                scoop_medium_middle()
            elif 0.4 <= weight_needed < 0.6:
                print("Using medium left scoop")
                scoop_medium_left()
            elif 0.3 <= weight_needed < 0.4:
                print("Using medium right scoop")
                scoop_medium_right()
            elif 0.2 <= weight_needed < 0.3:
                print("Using medium far left scoop")
                scoop_medium_farleft()
            elif 0.08 <= weight_needed < 0.2:
                print("Using micro scoop")
                scoop_tiny()
            elif 0.02 < weight_needed < 0.08:
                print("Using smallest micro scoop (random left/right)")
                scoop_fn = random.choice([scoop_tiny_left, scoop_tiny_right])
                scoop_fn()
            else:
                print("Remaining weight too small — consider nudging or final adjustment")

            for _ in range(30):
                raw_measurement = scale.measure()
                if raw_measurement.stable and abs(raw_measurement.value - previous_weight) > 0.01:
                    break
                time.sleep(0.1)

            weight_dispensed = raw_measurement.value
            print(f"Weight after scoop: {weight_dispensed} grams")
            weight_needed = calc_weight_needed(weight_dispensed, target_weight)
            print(f"Current weight needed: {weight_needed} grams")
            previous_weight = weight_dispensed

# Final result check
# Final result check
    difference = target_weight - weight_dispensed  # signed difference

    if difference < -0.02:
        print("⚠️ Overshot the target.")
        print("Final weight: {:.4f}g (Overshoot by {:.4f}g)".format(weight_dispensed, abs(difference)))
        replace_medium_spatula()
    elif -0.02 <= difference <= 0.02:
        print("✅ Target weight reached.")
        print("Final weight: {:.4f}g (Off by {:.4f}g)".format(weight_dispensed, abs(difference)))
        replace_medium_spatula()
    else:
        print("❓ Loop ended unexpectedly — undershot.")
        print("Final weight: {:.4f}g (Still need {:.4f}g to reach target)".format(weight_dispensed, difference))


```
