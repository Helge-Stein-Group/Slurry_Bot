```python
import time
import random
import datetime

# ── Safety limits ───────────────────────────────────────────
# If these are hit the container is probably empty
MAX_LARGE_SCOOPS  = 20
MAX_MEDIUM_SCOOPS = 30

# ── Logging ────────────────────────────────────────────────
log = []  # stores all events of the current run

def log_event(event_type, message, weight=None, added=None):
    """
    Logs any event during a weighing run.
    event_type: "scoop", "info", "warning", "error", "result"
    """
    entry = {
        "time":       datetime.datetime.now().strftime("%H:%M:%S"),
        "type":       event_type,
        "message":    message,
        "weight":     weight,   # total weight on scale at this point
        "added":      added,    # how much this scoop added (None for non-scoop events)
    }
    log.append(entry)

    # Print to notebook output as well
    weight_str = f" | total: {weight:.4f}g" if weight is not None else ""
    added_str  = f" | added: +{added:.4f}g" if added is not None else ""
    print(f"[{entry['time']}] [{event_type.upper():7s}] {message}{added_str}{weight_str}")

def print_run_summary():
    """Prints a full summary of the current run at the end."""
    print("\n" + "="*60)
    print("RUN SUMMARY")
    print("="*60)
    print(f"{'Time':<10} {'Type':<10} {'Event':<25} {'Added':>8} {'Total':>8}")
    print("-"*60)
    for entry in log:
        added_str  = f"{entry['added']:>8.4f}g" if entry['added']  is not None else f"{'':>9}"
        weight_str = f"{entry['weight']:>8.4f}g" if entry['weight'] is not None else f"{'':>9}"
        print(f"{entry['time']:<10} {entry['type'].upper():<10} {entry['message']:<25} {added_str} {weight_str}")
    print("="*60)

    # Count scoops and calculate stats
    scoops = [e for e in log if e["type"] == "scoop"]
    if scoops:
        total_added = sum(e["added"] for e in scoops if e["added"] is not None)
        print(f"Total scoops:     {len(scoops)}")
        print(f"Total dispensed:  {total_added:.4f}g")
    print("="*60)

# ── Weight measurement ──────────────────────────────────────
def calc_weight_needed(weight_dispensed, target_weight):
    """Returns how much weight is still needed to reach target."""
    return target_weight - weight_dispensed

def wait_for_stable_weight(previous_weight, max_attempts=30):
    """
    Waits until the scale returns a stable reading.
    Raises TimeoutError if scale doesn't stabilise in time.
    """
    for _ in range(max_attempts):
        raw = scale.measure()
        if raw.stable and abs(raw.value - previous_weight) > 0.01:
            return raw.value
        time.sleep(0.1)
    raise TimeoutError("Scale did not stabilise – check scale connection!")
```

```python
# ============================================================
# CALIBRATION – run before first real weighing session!
# Tests each scooping motion N times and prints average yield
# Compare results with Aidan's values in the report to check
# if coordinates are still correct after the shelf was moved
# ============================================================

def calibrate_scoop(scoop_function, name, repetitions=5):
    print(f"\n--- Calibrating: {name} ({repetitions}x) ---")
    print("Zeroing scale...")
    scale.tare()

    results = []
    previous_weight = 0

    for i in range(repetitions):
        scoop_function()
        weight = wait_for_stable_weight(previous_weight)
        added = weight - previous_weight
        results.append(added)
        print(f"  Scoop {i+1}: +{added:.4f}g (total: {weight:.4f}g)")
        previous_weight = weight

    avg = sum(results) / len(results)
    print(f"  → Average: {avg:.4f}g | Min: {min(results):.4f}g | Max: {max(results):.4f}g")
    return avg

# Comment out any functions you don't want to test

# Large spatula – expected from Aidan's report: middle ~1.0g, left ~0.6g, right ~0.5g
calibrate_scoop(scoop_big_middle,     "big middle",    repetitions=3)
calibrate_scoop(scoop_big_left,       "big left",      repetitions=3)
calibrate_scoop(scoop_big_right,      "big right",     repetitions=3)

# Medium spatula – expected: left ~0.3g, middle ~0.23g, right ~0.21g, farleft ~0.215g
calibrate_scoop(scoop_medium_left,    "medium left",    repetitions=3)
calibrate_scoop(scoop_medium_middle,  "medium middle",  repetitions=3)
calibrate_scoop(scoop_medium_right,   "medium right",   repetitions=3)
calibrate_scoop(scoop_medium_farleft, "medium farleft", repetitions=3)

# Tiny scoops – expected: ~0.05-0.08g range
calibrate_scoop(scoop_tiny,       "tiny middle", repetitions=3)
calibrate_scoop(scoop_tiny_left,  "tiny left",   repetitions=3)
calibrate_scoop(scoop_tiny_right, "tiny right",  repetitions=3)
```

```python
def run_weighing_logic(target_weight):
    """
    Internal function – contains all scooping and weighing logic.
    Called by run_weighing_workflow(), not directly.
    
    Requires:
    - scale to be connected and tared beforehand
    - robot to be connected and initialized
    - all scoop functions to be defined (cells 6 & 7)
    - log and wait_for_stable_weight to be defined (cell 4)
    """

    global previous_weight, last_scoop

    previous_weight = 0
    last_scoop      = None

    # ── Select and pick up spatula ──────────────────────────
    if target_weight >= 1.0:
        pickup_large()
        log_event("info", "Large spatula selected and picked up")
    else:
        pickup_medium()
        log_event("info", "Medium spatula selected and picked up")

    # ── Read starting weight ────────────────────────────────
    weight_dispensed = scale.measure_stable().value
    weight_needed    = calc_weight_needed(weight_dispensed, target_weight)
    log_event("info", f"Starting weight: {weight_dispensed}g | Needed: {weight_needed}g")

    # ============================================================
    # CASE A: target >= 1g → large spatula first, then medium
    # ============================================================
    if target_weight >= 1.0:

        large_scoop_count = 0

        # ── First scoop based on target weight ─────────────
        if 1.0 <= target_weight < 1.5:
            scoop_big_right()
            last_scoop = scoop_big_right
            log_event("scoop", "big right (first scoop)")
        elif 1.5 <= target_weight < 2.0:
            scoop_big_left()
            last_scoop = scoop_big_left
            log_event("scoop", "big left (first scoop)")
        else:
            scoop_big_middle()
            last_scoop = scoop_big_middle
            log_event("scoop", "big middle (first scoop)")
        large_scoop_count += 1

        # ── Measure after first scoop ───────────────────────
        weight_dispensed = wait_for_stable_weight(previous_weight)
        weight_needed    = calc_weight_needed(weight_dispensed, target_weight)
        fraction_left    = weight_needed / target_weight
        log_event("scoop", "first scoop measured",
                  weight=weight_dispensed,
                  added=weight_dispensed - previous_weight)
        previous_weight = weight_dispensed

        # ── Large spatula loop ──────────────────────────────
        # Keep scooping until only 37% of target weight is remaining
        # Why 37%? → remainder is then small enough for medium spatula
        while fraction_left > 0.37:

            # Safety check – stop if too many scoops
            if large_scoop_count >= MAX_LARGE_SCOOPS:
                log_event("warning",
                          f"Large spatula limit reached ({MAX_LARGE_SCOOPS}) – container probably empty!")
                print("⚠️ Stopping large spatula loop – check if container is empty!")
                break

            if fraction_left > 0.62:
                scoop_big_middle()
                last_scoop = scoop_big_middle
                log_event("scoop", "big middle")
            elif 0.5 <= fraction_left <= 0.62:
                scoop_big_left()
                last_scoop = scoop_big_left
                log_event("scoop", "big left")
            elif 0.4 <= fraction_left < 0.5:
                # Second scoop from same spot reliably gives ~half the amount
                last_scoop()
                log_event("scoop", "repeat last scoop")
            elif 0.37 <= fraction_left < 0.4:
                scoop_big_right()
                last_scoop = scoop_big_right
                log_event("scoop", "big right")
            else:
                log_event("error", "Unexpected fraction state in large spatula loop!")
                break

            large_scoop_count += 1

            weight_dispensed = wait_for_stable_weight(previous_weight)
            weight_needed    = calc_weight_needed(weight_dispensed, target_weight)
            fraction_left    = weight_needed / target_weight
            log_event("scoop", f"large scoop #{large_scoop_count}",
                      weight=weight_dispensed,
                      added=weight_dispensed - previous_weight)
            previous_weight = weight_dispensed

        # ── Switch to medium spatula ────────────────────────
        if 0.03 <= fraction_left < 0.37:
            log_event("info", "Switching to medium spatula")
            replace_big_spatula()
            pickup_medium()

            medium_scoop_count = 0

            while weight_needed > 0.05:

                # Safety check
                if medium_scoop_count >= MAX_MEDIUM_SCOOPS:
                    log_event("warning",
                              f"Medium spatula limit reached ({MAX_MEDIUM_SCOOPS}) – container probably empty!")
                    print("⚠️ Stopping medium spatula loop – check if container is empty!")
                    break

                if weight_needed >= 0.6:
                    scoop_medium_left()
                    log_event("scoop", "medium left")
                elif 0.4 <= weight_needed < 0.6:
                    scoop_medium_middle()
                    log_event("scoop", "medium middle")
                elif 0.3 <= weight_needed < 0.4:
                    scoop_medium_right()
                    log_event("scoop", "medium right")
                elif 0.2 <= weight_needed < 0.3:
                    scoop_medium_farleft()
                    log_event("scoop", "medium farleft")
                elif 0.08 <= weight_needed < 0.2:
                    scoop_tiny_right()
                    log_event("scoop", "tiny right")
                elif 0.04 <= weight_needed < 0.08:
                    scoop_tiny_left()
                    log_event("scoop", "tiny left")
                else:
                    log_event("warning", "Remainder too small – manual adjustment needed")
                    print("⚠️ Remainder too small for automatic scooping")
                    break

                medium_scoop_count += 1

                weight_dispensed = wait_for_stable_weight(previous_weight)
                weight_needed    = calc_weight_needed(weight_dispensed, target_weight)
                log_event("scoop", f"medium scoop #{medium_scoop_count}",
                          weight=weight_dispensed,
                          added=weight_dispensed - previous_weight)
                previous_weight = weight_dispensed

            # ── Final result Case A ─────────────────────────
            difference = target_weight - weight_dispensed
            if difference < -0.02:
                log_event("result",
                          f"OVERSHOT by {abs(difference):.4f}g",
                          weight=weight_dispensed)
                print(f"⚠️ Overshot! {weight_dispensed:.4f}g (by {abs(difference):.4f}g)")
            elif -0.02 <= difference <= 0.02:
                log_event("result",
                          f"TARGET REACHED, off by {abs(difference):.4f}g",
                          weight=weight_dispensed)
                print(f"✅ Done! {weight_dispensed:.4f}g (off by {abs(difference):.4f}g)")
            else:
                log_event("result",
                          f"UNDERSHOT by {difference:.4f}g",
                          weight=weight_dispensed)
                print(f"❓ Undershot. {weight_dispensed:.4f}g ({difference:.4f}g still needed)")

            replace_medium_spatula()

        else:
            log_event("error",
                      f"Fraction {fraction_left:.4f} outside expected range after large spatula loop")
            print("Error: fraction outside expected range – check container level")

    # ============================================================
    # CASE B: target < 1g → medium spatula only
    # ============================================================
    else:
        medium_scoop_count = 0

        while weight_needed > 0.02:

            # Safety check
            if medium_scoop_count >= MAX_MEDIUM_SCOOPS:
                log_event("warning",
                          f"Medium spatula limit reached ({MAX_MEDIUM_SCOOPS}) – container probably empty!")
                print("⚠️ Stopping – check if container is empty!")
                break

            if weight_needed >= 0.6:
                scoop_medium_left()
                log_event("scoop", "medium left")
            elif 0.4 <= weight_needed < 0.6:
                scoop_medium_middle()
                log_event("scoop", "medium middle")
            elif 0.3 <= weight_needed < 0.4:
                scoop_medium_right()
                log_event("scoop", "medium right")
            elif 0.2 <= weight_needed < 0.3:
                scoop_medium_farleft()
                log_event("scoop", "medium farleft")
            elif 0.08 <= weight_needed < 0.2:
                scoop_tiny()
                log_event("scoop", "tiny middle")
            elif 0.02 < weight_needed < 0.08:
                # Random choice prevents infinite loop if one area is empty
                scoop_fn = random.choice([scoop_tiny_left, scoop_tiny_right])
                log_event("scoop", f"tiny random ({scoop_fn.__name__})")
                scoop_fn()
            else:
                log_event("warning", "Remainder too small – manual adjustment needed")
                print("⚠️ Remainder too small for automatic scooping")
                break

            medium_scoop_count += 1

            weight_dispensed = wait_for_stable_weight(previous_weight)
            weight_needed    = calc_weight_needed(weight_dispensed, target_weight)
            log_event("scoop", f"medium scoop #{medium_scoop_count}",
                      weight=weight_dispensed,
                      added=weight_dispensed - previous_weight)
            previous_weight = weight_dispensed

        # ── Final result Case B ─────────────────────────────
        difference = target_weight - weight_dispensed
        if difference < -0.02:
            log_event("result",
                      f"OVERSHOT by {abs(difference):.4f}g",
                      weight=weight_dispensed)
            print(f"⚠️ Overshot! {weight_dispensed:.4f}g (by {abs(difference):.4f}g)")
        elif -0.02 <= difference <= 0.02:
            log_event("result",
                      f"TARGET REACHED, off by {abs(difference):.4f}g",
                      weight=weight_dispensed)
            print(f"✅ Done! {weight_dispensed:.4f}g (off by {abs(difference):.4f}g)")
        else:
            log_event("result",
                      f"UNDERSHOT by {difference:.4f}g",
                      weight=weight_dispensed)
            print(f"❓ Undershot. {weight_dispensed:.4f}g ({difference:.4f}g still needed)")

        replace_medium_spatula()
```

```python
# CELL – Full Weighing Workflow
# ============================================================
# Combines Robot_wrapper functions with weighing logic
# ============================================================

def weighing_workflow(vial_number, target_weight):
    """
    Full workflow:
    1. Pick up vial and place on scale    (Robot_wrapper)
    2. Place funnel over vial             (Aidans code)
    3. Pick up spatula and weigh          (Aidans code)
    4. Replace spatula and funnel         (Aidans code)
    5. Put vial back                      (Robot_wrapper) ← noch nicht fertig
    """

    log.clear()
    log_event("info", f"Starting workflow | Vial: {vial_number} | Target: {target_weight}g")

    # ── Step 1: Pick up vial and place on scale ─────────────
    log_event("info", f"Picking up {vial_number} and moving to scale")
    robot.PickUpVial(vial_number)   # aus Robot_wrapper!

    # ── Step 2: Place funnel ────────────────────────────────
    log_event("info", "Placing funnel")
    place_funnel()                  # aus Aidans Code

    # ── Step 3: Weigh ───────────────────────────────────────
    log_event("info", "Starting weighing sequence")
    run_weighing_logic(target_weight)   # die komplette Wiegelogik als eigene Funktion

    # ── Step 4: Replace funnel ──────────────────────────────
    log_event("info", "Replacing funnel")
    replace_funnel()

    # ── Step 5: Put vial back ───────────────────────────────
    # TODO: noch nicht implementiert, muss noch in Robot_wrapper ergänzt werden
    log_event("info", "Vial return not yet implemented")
    print("⚠️ Remember to manually remove vial from scale!")

    log_event("info", "Workflow complete")
    print_run_summary()
```
